# Word2Vec model using Google pre trained embedding weights and required funtions

```python

def _removeNonAscii(s):
    return "".join(i for i in s if  ord(i)<128)

def make_lower_case(text):
    return text.lower()

def remove_stop_words(text):
    text = text.split()
    stops = set(stopwords.words("english"))
    text = [w for w in text if not w in stops]
    text = " ".join(text)
    return text

def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)

def remove_punctuation(text):
    tokenizer = RegexpTokenizer(r'\w+')
    text = tokenizer.tokenize(text)
    text = " ".join(text)
    return text
```

```python
global word_embeddings
word_embeddings = []

for line in dfCleaned['cleaned']:
    avgword2vec = None
    count = 0
    for word in line.split():
        if word in google_model.wv.vocab:
            count += 1
            if avgword2vec is None:
                avgword2vec = google_model[word]
            else:
                avgword2vec = avgword2vec + google_model[word]
            
    if avgword2vec is not None:
        avgword2vec = avgword2vec / count
    
        word_embeddings.append(avgword2vec)
```

```python

def vectors2(x):
    
    global input_embeddings
    input_embeddings = []


    avgword2vec = None
    count = 0
    for word in x.split():
        if word in google_model.wv.vocab:
            count += 1
            if avgword2vec is None:
                avgword2vec = google_model[word]
            else:
                avgword2vec = avgword2vec + google_model[word]
            
    if avgword2vec is not None:
        avgword2vec = avgword2vec / count
    
    input_embeddings.append(avgword2vec)
    return input_embeddings
```

```python
def Nmaxelements(list1, N):
    final_list = []
    index_=[]
    
    for i in range(0, N): 
        max1 = 0
        index=-1      
        for j in range(len(list1)):     
            if list1[j] > max1:
                max1 = list1[j];
                index=j  
        list1.remove(max1);
        index_.append(index)
        final_list.append(max1)
          
    return index_
```

## Average Word2Vec method for sentecnce

```python
def recommendAVGword2Vec(str,N):
    cosine_similarities = cosine_similarity(word_embeddings,vectors2(str))
    score=[]
    for i in range(0,len(cosine_similarities)):
        score.append(cosine_similarities[i][0])
    index=Nmaxelements(score,N)
    for i in index:
        print(df["product_info"][i])
        print(df["desc"][i])
        print("----")

```

```python
recommendAVGword2Vec("baby products",5)
```

## CNN SEARCH ENGINE

```python
from keras.models import model_from_json

json_file = open('models/model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("models/weights_cnn_sentece.hdf5")
print("Loaded model from disk")
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

dfCluster=pd.read_csv("productInfoClusters.csv")
vocab_size = 20000
tokenizer = Tokenizer(num_words=vocab_size) 
```

```python
def CNN(str1):
    s=tokenizer.texts_to_sequences([str1])
    input=pad_sequences(s, maxlen=100)
    # Get prediction
    pred = loaded_model.predict(input[0].reshape(1,100))
    clsuterNo=np.argmax(pred)
    df2=dfCluster[dfCluster["cluster"]==clsuterNo]
    print(df2["product_info"])
```

```python
str1="weight loss"
CNN("str1")
```

```python
def main(str,N):
    print("Kmeans with Google pretrained")
    kmeansGoogle(str)
    print("==========================")
    print("TDIDF Word2Vec")
    recommendTDIDF(str,N)
    print("==========================")
    print("Kmeans with tfidf")
    kmeansTFIDF(str)
    print("==========================")
    print("Average Word2Vec")
    recommendAVGword2Vec(str,N)
```
